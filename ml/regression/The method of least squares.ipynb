# The Method of Least Squares

[Source](https://web.williams.edu/Mathematics/sjmiller/public_html/BrownClasses/54/handouts/MethodLeastSquares.pdf)

Given data *{(x<sub>1</sub>, y<sub>1</sub>), ..., (x<sub>N</sub>, y<sub>N</sub>)}* we may define the error assaciated to saying *y = ax + b* by

```python
%%latex

$$E(a, b) = \sum{(y_n - (ax_n + b)^2)}$$
```

The goal is to find values of ``a`` and ``b`` that minimize the error.
